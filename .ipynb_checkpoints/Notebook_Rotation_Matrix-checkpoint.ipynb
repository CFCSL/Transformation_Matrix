# Rotation Matrix




## direct approach using Numpy:

```python
ndarray=np.array([[1,2,3], [4, 5,6]])
display(ndarray)
angles=[0,0,30]
r_matrix=rotation_matrix(angles)
display(r_matrix)

rotated_matrix=apply_rotation(ndarray, r_matrix) # or simply use a dot product: np.dot(ndarray, r_matrix)
display(rotated_matrix)
```
